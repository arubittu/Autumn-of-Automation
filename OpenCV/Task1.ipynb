```python
cv2.imwrite(r'C:\Users\ARYAN P. KOLAPKAR\Downloads\starry.jpg',image)
```

```python
gray=cv2.cvtColor(image,cv2.cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray,cmap='gray') #grayscale
```

```python
ret,bw = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)
plt.imshow(bw,cmap='gray') # black and white 
```

```python
thresh1 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, 
                                          cv2.THRESH_BINARY, 199, 5) 
plt.imshow(thresh1,cmap='gray') # adaptive threshold
```

```python
image=cv2.cvtColor(image,cv2.COLOR_RGB2BGR)
```

```python
cv2.imshow('',image)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
cap=cv2.VideoCapture(0)
while True:
    ret,frame=cap.read()
    cv2.imshow('img', frame)
    k = cv2.waitKey(2) & 0xff
    if k==27:
        cap.release()
        cv2.destroyAllWindows()
        break
```
